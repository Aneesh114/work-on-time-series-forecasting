```python
def impute(data):
    dataframe = pd.DataFrame(data)
    dataframe = dataframe.iloc[:,:].interpolate(method="polynomial", order=2)
    dataframe = dataframe.fillna(method='ffill')
    dataframe = dataframe.fillna(method='bfill')
    return dataframe
```

```python
def scale(data):
    dataframe = pd.DataFrame(data)
    from sklearn.preprocessing import StandardScaler
    scale= StandardScaler()
    scaled_data = scale.fit_transform(data.iloc[:,:]) 
    return scaled_data
    
```

```python
def convert_datetime(data,datetime_column):
    from datetime import datetime
    dataframe = pd.DataFrame(data)
    dataframe[datetime_column] = pd.to_datetime(dataframe[datetime_column])
    z = []
    for i in np.arange(0,len(dataframe)):
        z.append(int(dataframe[datetime_column][i].timestamp()))
    dataframe['timestamp'] = z
    z = []
    for i in np.arange(0,len(dataframe)):
        z.append(datetime.fromtimestamp(dataframe['timestamp'][i]))
    dataframe['new_datetime'] = z
    return dataframe
    
```
