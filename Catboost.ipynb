```python
# importing the dataset
X_train = pd.DataFrame(np.loadtxt("Data/UCI HAR Dataset/train/X_train.txt"))
y_train = pd.DataFrame(np.loadtxt("Data/UCI HAR Dataset/train/y_train.txt"))
X_test = pd.DataFrame(np.loadtxt("Data/UCI HAR Dataset/test/X_test.txt"))
y_test = pd.DataFrame(np.loadtxt("Data/UCI HAR Dataset/test/y_test.txt"))
```

```python
# doing pre-processing of the train and test dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#pip install catboost
X_train
```

```python
# catboost classifier
cat_cv_classifier = CatBoostClassifier(
    iterations = 1000, # 1000 are ideal
    loss_function='MultiClass',
    bootstrap_type = "Bayesian",
    eval_metric = 'MultiClass',
    leaf_estimation_iterations = 100,
    random_strength = 0.5,
    depth = 7,
    l2_leaf_reg = 5,
    learning_rate=0.1,
    bagging_temperature = 0.5,
    task_type = "GPU",
)
```

```python
# cross validation function to find the best parameters
scores = []
for i in range(10):
    scores.append(cross_validate(cat_cv_classifier,X_train,y_train))
print(scores)
print("The 10 times 10 fold accuracy of the catboost classifier is: ",np.average(scores))    
```
